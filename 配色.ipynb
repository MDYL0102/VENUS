```
library(ggplot2)

df<-data.frame(Blocks=factor(rep(letters[1:10],each=5)),GamesPlayed=rnorm(50))
ggplot(df,aes(x=Blocks,y=GamesPlayed))+geom_point(color=4)
#x=火鍋數量 Y=出場數 求場均火鍋
```

```
library(ggplot2)

df<-data.frame(Steals=factor(rep(letters[1:10],each=10)),GamesPlayed=rnorm(100))

ggplot(df,aes(x=Steals,y=GamesPlayed))+geom_boxplot(fill=4)
#x=抄截數量 Y=出場數 求場均抄截
```
